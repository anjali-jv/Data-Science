Creating `rating` column from `name` column by pulling all numbers after the `★` character.

```python
# Do NOT modify this code cell

# creating 'rating' column
star_char = "★"

df['rating'] = df['name'].apply(lambda x: re.findall(star_char + "(\\d+.?\\d*)", x)[0] if len(re.findall(star_char + "(\\d+.?\\d*)", x)) > 0 else np.nan)
df.head()
```

```python
# your code start here #
log_price = np.log(df['price'])
df['z-score_of_log_price'] = (log_price - log_price.mean()) / log_price.std()
# end #
df['z-score_of_log_price'].head(5)
```

## Q4(c)

```python
cols_to_output = ["id", "name", "host_id", "host_name", "neighbourhood", "room_type", "price", "number_of_reviews", "rating"]

zscore_threshold = 3

# your code start here #
top_priced_listings = df[df['z-score_of_log_price'] < -zscore_threshold][cols_to_output]
top_priced_listings = top_priced_listings.sort_values(by='price', ascending=False)
# end #

print(f"The shape of top_priced_listings is: {top_priced_listings.shape}")

top_priced_listings # Outputs top_priced_listings.
                    # Note, Google Colab may truncate this outputs to the first five and last five rows. If done, leave your output truncated by Google Colab.
```

## Q4(d)

```python
# Do NOT modify this code cell
modified_copy = df.copy()
print(f"The shape of data frame is: {modified_copy.shape}")
```

```python
# your code start here #
quartile_1 = np.nanpercentile(modified_copy['price'].astype(float), 25)
quartile_3 = np.nanpercentile(modified_copy['price'].astype(float), 75)
IQR = quartile_3 - quartile_1

lower_bound = quartile_1 - 1.5 * IQR
upper_bound = quartile_3 + 1.5 * IQR

modified_copy = modified_copy[(modified_copy['price'] >= lower_bound) & (modified_copy['price'] <= upper_bound)]

# end #

print(f"The shape of data frame is: {modified_copy.shape}")
```
